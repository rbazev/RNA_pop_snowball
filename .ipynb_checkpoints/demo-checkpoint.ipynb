# The Russian roulette model

```python
%run RR_model.py
```

The ``Population`` object is initialised by the following parameters:

- `pop_size` : Population size

- `L` : length of the binary sequence

- `p_net` : probability of viability

- `U_rate` : genomic muatation rate

- `r_prob` : the recombination probability; offspring are recombinant resulting from a single crossover. If set to `Free`, Crossovers488
occur independently with probability $0.5$ at each of the $L − 1$ sites. 

```python
pop = Population(pop_size=50, L=30, p_net=0.5, U_rate=0.1, r_prob=0.0)
twopop = TwoPops(pop, t=1000)
twopop.evolve_burnin()
output = twopop.diverge(target_D=0.1, tot_sample=5)
```

```python
fig, ax = plt.subplots(
    nrows=1,
    ncols=1,
    figsize=(5, 5), constrained_layout=True)

plt.scatter(output['D'], output['II'])
plt.xlabel('Genetic distance')
plt.ylabel('Incompatible introgressions');
```
